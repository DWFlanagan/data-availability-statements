```python
import pandas as pd
import jaydebeapi
import plotly.express as px
import pandas_profiling

import os
from dotenv import load_dotenv, find_dotenv
load_dotenv(find_dotenv())
```

```python
NETEZZA_URL = os.getenv("NETEZZA_URL")
NETEZZA_USER = os.getenv("NETEZZA_USER")
NETEZZA_PASSWORD = os.getenv("NETEZZA_PASSWORD")

jdbc_driver_loc = "nzjdbc.jar"
jdbc_driver_name = "org.netezza.Driver"
connection_string = f"jdbc:netezza://{NETEZZA_URL}/"
```

```python
# Connect to Netezza

conn = jaydebeapi.connect(
    "org.netezza.Driver",
    connection_string + "SUBMISSIONS",
    {"user": NETEZZA_USER, "password": NETEZZA_PASSWORD},
    jars=jdbc_driver_loc,
)
```

```python
stmt = """
SELECT SITE_NAME,
    DOCUMENTID,
    CUSTOMQUESTIONID,
    QUESTIONTEXT,
    SUBMISSIONCUSTOMANSWER_ANSWERTEXT AS answer_text
    
FROM S1_ARTICLES_SUBMISSION_CUSTOM_QUESTIONS

-- Looking for any question or answer that mentions Data, since there may be relevant questions besides "Wiley Data Availability Statement.
-- Using UPPER() since Netezza is case sensitive and doesn't have ILIKE.

WHERE
    UPPER(QUESTIONTEXT) LIKE '%DATA%'
    OR UPPER(ANSWER_TEXT) LIKE '%DATA%'
    
"""

df = pd.read_sql(stmt, conn)

df.head()
```

```python
df.profile_report()
```

```python
df.to_csv("statements.csv", index=False, sep="|")
```
